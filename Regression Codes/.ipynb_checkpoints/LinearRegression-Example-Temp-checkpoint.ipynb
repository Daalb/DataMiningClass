# Simple Linear Regression

The housing dataset contains information about different houses in Boston. Data is available from the scikit-learn library. The objective is to predict the value of prices of the house using the given features.

```python
#import libraries and read dataset
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.model_selection
from sklearn.linear_model import LinearRegression

boston = pd.read_csv('boston.csv')
boston.head(5)

#Boston dataset variables
#medv: median house value
#rm: average number pf rooms per dwelling
```

```python
#Plotting the dataset

plt.scatter(boston.rm,boston.medv)
plt.xlabel("Avg. Number of Rooms per Dwelling: $X_i$")
plt.ylabel("Median House Value: $Y_i$")
plt.title("Avg. Number of Rooms vs Median House Value")
plt.show()
```

```python
# Split dataset into target values Y and predictor values X

X = boston.drop('medv',axis=1)#Solo me qued√© con la columna rm
Y = boston['medv']
```

```python
# Split the dataset into train and test data
X_train, X_test, Y_train, Y_test = sklearn.model_selection.train_test_split(X, Y, test_size =0.33 , random_state = 5)
#Me divide el conjunto de datos entre train y test//El test_size depende de la representaci√≥n muestral, si es muy buena bajo el
#tama√±o del testeo//random_state es un valor para escoger de manera aleatoria los datos

# Run linear regression
lm = LinearRegression()#Inicializo el modelo 
lm.fit(X_train,Y_train)#Ajusto el modelo con los conjuntos de datos de entrenamiento

Y_pred = lm.predict(X_test)#Cuales son mis predicciones para Y


plt.scatter(X,Y)#Todo el conjunto de puntos muestrales
plt.plot(X_test,Y_pred,'red')#Mi modelo
plt.xlabel("Avg. Number of Rooms per Dwelling: $X_i$")
plt.ylabel("Median House Value: $Y_i$")
plt.title("Avg. Number of Rooms vs Median House Value")
plt.show()

#En este caso la regresi√≥n lineal est√° definiendo el conjunto de datos, si se ajust√≥ bien
```

```python
# Mean squared error
mse = sklearn.metrics.mean_squared_error(Y_test, Y_pred)
print(mse)
```

```python
# R-square
lm.score(X_test,Y_test)
```

```python
print('intercept:',lm.intercept_ )
print('slope:',lm.coef_ )
```
