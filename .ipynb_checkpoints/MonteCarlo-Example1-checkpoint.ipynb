```python
import numpy as np
from scipy.stats import uniform

N=1000
T=0

for i in range(N):
    x = uniform.rvs(size=10)
    h=0
    for xi in x:
        if  xi<0.5:
            h+=1
    
    if h==3 or h==6 or h==9:
        T+=1

        
print(T/N)
```

```python
#Simulo los lanzamientos
def lanzamiento():
    num = random.randint(1,100)
    if 1<=num<=51:
        return False;
    else:
        return True;
```

```python
def apuesta(x,dinero,monto):
    montos = []
    for i in x:
        if lanzamiento():
            dinero = dinero + monto
            montos.append(dinero)
        else:
            dinero = dinero - monto
            montos.append(dinero)
    
    plt.plot(x,montos)   
```

```python
num_simulaciones = 100
x = np.mgrid[1:20:100j]
for i in range(num_simulaciones):
    apuesta(x,10000,100)
    
```
